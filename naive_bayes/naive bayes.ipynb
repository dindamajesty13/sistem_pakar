```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
```

```python
df = pd.read_csv("heart.csv", sep=',')
```

```python
df
```

```python
encode = LabelEncoder()
df['age'] = encode.fit_transform(df['age'])
df['sex'] = encode.fit_transform(df['sex'])
df['cp'] = encode.fit_transform(df['cp'])
df['trestbps'] = encode.fit_transform(df['trestbps'])
df['chol'] = encode.fit_transform(df['chol'])
df['fbs'] = encode.fit_transform(df['fbs'])
df['restecg'] = encode.fit_transform(df['restecg'])
df['thalach'] = encode.fit_transform(df['thalach'])
df['exang'] = encode.fit_transform(df['exang'])
df['oldpeak'] = encode.fit_transform(df['oldpeak'])
df['slope'] = encode.fit_transform(df['slope'])
df['ca'] = encode.fit_transform(df['ca'])
df['thal'] = encode.fit_transform(df['thal'])
df['target'] = encode.fit_transform(df['target'])
```

```python
x = df.drop(["target"], axis = 1)
x.head()
```

```python
y = df["target"]
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 100)
```

```python
nb = GaussianNB()
```

```python
nb_train = nb.fit(x_train, y_train)
```

```python
y_pred = nb_train.predict(x_test)
y_pred
```

```python
np.array(y_test)
```

```python
nb_train.predict_proba(x_test)
```

```python
confusion_matrix(y_test, y_pred)
f1_score(y_test, y_pred, labels=np.unique(y_pred), average='weighted')
```

```python
classification_report(y_test, y_pred)
```
